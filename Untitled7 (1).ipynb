```python
import pandas as pd
import numpy as np
```

```python
train_dir = '/content/drive/MyDrive/OA/TRAIN'
test_dir = '/content/drive/MyDrive/OA/TEST'
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
train_data = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)
```

```python
test_datagen = ImageDataGenerator(rescale=1.0/255)
test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)
```

```python
import matplotlib.pyplot as plt
import numpy as np


images, labels = next(train_data)


random_index = np.random.randint(0, len(images))
image = images[random_index]
label = labels[random_index]

class_name = "caries" if label == 0 else "gingivitis"

plt.imshow(image)
plt.title(f"Label: {class_name}")
plt.axis('off')
plt.show()
```

```python
from collections import Counter


train_class_counts = Counter(train_data.classes)
print("Training Data:")
print(f"Caries: {train_class_counts[0]}")
print(f"Gingivitis: {train_class_counts[1]}")


test_class_counts = Counter(test_data.classes)
print("\nTesting Data:")
print(f"Caries: {test_class_counts[0]}")
print(f"Gingivitis: {test_class_counts[1]}")
```

```python

print(f"Number of images in training data: {train_data.n}")
print(f"Number of images in test data: {test_data.n}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization


model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])


model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_data,
    epochs=10,
    validation_data=test_data
)
```

```python
loss, accuracy = model.evaluate(test_data)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt

def classify_single_image(image_path, model, class_labels=['caries', 'gingivitis']):

    image = load_img(image_path, target_size=(224, 224))


    image_array = img_to_array(image) / 255.0


    image_array = np.expand_dims(image_array, axis=0)


    probability = model.predict(image_array)[0][0]

    predicted_label = class_labels[1] if probability >= 0.5 else class_labels[0]


    plt.imshow(image)
    plt.title(f"Prediction: {predicted_label} ({probability:.2f})")
    plt.axis('off')
    plt.show()

    return predicted_label


image_path = '/content/drive/MyDrive/OA/TEST/Caries/197_1.jpg'

predicted_class = classify_single_image(image_path, model)
print(f"The predicted class for the input image is: {predicted_class}")
```
